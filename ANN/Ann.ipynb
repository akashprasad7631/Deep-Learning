```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Flatten, Dense
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/akjadon/Deep-Learning/master/ANN/churn_ann_model/Customer_Churn_Modelling.csv")
```

```python
df.head()
```

```python
df = df.drop(['CustomerId', 'Surname', 'RowNumber'], axis = 1)
```

```python
df.head()
```

```python
df.info()
```

```python
from sklearn.preprocessing import LabelEncoder
label1 = LabelEncoder()
df["Geography"] = label1.fit_transform(df["Geography"]) 
```

```python
df["Gender"] = label1.fit_transform(df["Gender"])
```

```python
df.head()
```

```python
df["Geography"].value_counts()
```

```python
x = df.drop(['Exited'], axis = 1)
y = df['Exited']
```

```python
x.head()
```

```python
y.head()
```

```python
x = pd.get_dummies(x, drop_first=True, columns=['Geography'])
x.head()
```

```python
from sklearn.preprocessing import StandardScaler
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 0, stratify = y)
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
model = Sequential()
model.add(Dense(x.shape[1],input_dim=x.shape[1],activation='relu'))
model.add(Dense(216,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
```

```python
model.compile(optimizer='adam', loss = 'binary_crossentropy', metrics=['accuracy'])
```

```python
model.fit(x_train, y_train.to_numpy(), batch_size = 10, epochs = 10, verbose = 1,validation_data=(x_test,y_test.to_numpy()))
```
